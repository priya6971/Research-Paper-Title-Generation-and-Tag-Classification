```python
import numpy as np 
import pandas as pd 

from sklearn.metrics import f1_score, accuracy_score,confusion_matrix 
```

```python
def Load_TrueLabels_PredLabels(true_lab,pred_lab):
  """
  Gets the input lists of path of true label arrays of each class and prediction label list path of each class
  Input : true_lab, pred_lab - path list for each class 
  Returns : 5-dimensional true label and predicted label numpy arrays 
  """
  original_labels=np.load(true_lab[0],allow_pickle=True).reshape(-1,1)
  predicted_labels=np.argmax(np.load(pred_lab[0],allow_pickle=True),axis=1).flatten().reshape(-1,1)
  for true_path,pred_path in zip(true_lab[1:],pred_lab[1:]):
    original_labels=np.concatenate([original_labels,np.load(true_path,allow_pickle=True).reshape(-1,1)],axis=1)
    predicted_labels=np.concatenate([predicted_labels,np.argmax(np.load(pred_path,allow_pickle=True),axis=1).flatten().reshape(-1,1)],axis=1)
  
  return original_labels, predicted_labels
```

```python
test_true_label_path=["/content/AI_test_true_labels.npy","/content/CL_test_true_labels.npy","/content/CV_test_true_labels.npy","/content/LG_test_true_labels.npy","/content/NE_test_true_labels.npy"]
test_pred_label_path=["/content/AI_test_pred.npy","/content/CL_test_pred.npy","/content/CV_test_pred.npy","/content/LG_test_pred.npy","/content/NE_test_pred.npy"]

true_labels, pred_labels = Load_TrueLabels_PredLabels(test_true_label_path, test_pred_label_path)
```

```python
accuracy_score(true_labels,pred_labels)
```

```python
f1_score(true_labels,pred_labels,average="micro")
```
